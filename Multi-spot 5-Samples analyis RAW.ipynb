##FRET fitting remarks
Three fitting models defined, 1-Gaussian and 2-Gaussian mixture and 2-Gaussian with free amplitudes.

* 1-Gaussian model (**1gauss**):

$$f_{PDF}(x, \mu, \sigma) = \frac{1}{\sigma\sqrt{2\pi}} exp\left( -\frac{(x-\mu)^2}{2\sigma^2} \right)$$

* 2-Gaussian mixture model (**2gauss**):

$$f_{2-PDF}(x, \mu_1, \sigma_1, \mu_2, \sigma_2, \alpha) = \alpha f_{PDF}(x, \mu_1, \sigma_1) + (1-\alpha)f_{PDF}(x, \mu_2, \sigma_2)$$

* 2-Gaussian model with free amplitudes (**2gauss_fun**):

$$f_{2-PDF}(x, \mu_1, \sigma_1, \alpha_1, \mu_2, \sigma_2, \alpha_2) = \alpha_1 f_{PDF}(x, \mu_1, \sigma_1) + \alpha_2 f_{PDF}(x, \mu_2, \sigma_2)$$

The model is chosen for each measurements setting the `fit_type` variable.

##Utility functions

```python
## Define some utility functions

def print_fit(d, gamma_post, math=True):
    """Print fit and standard deviation for both corrected and uncorrected E
    Returns d.E_fit.
    """
    E_fit = d.E_fit
    Ec = bl.gamma_correct_E(E_fit, gamma_post)
    if math:
        display(Math(r'\text{Pre}\;\gamma\quad\langle{E}_{fit}\rangle = %.1f\%% \qquad'
                     '\Delta E_{fit} = %.2f \%%' % \
                     (E_fit.mean()*100,(E_fit.max()-E_fit.min())*100)))
        display(Math(r'\text{Post}\;\gamma\quad\langle{E}_{fit}\rangle = %.1f\%% \qquad'
                     '\Delta E_{fit} = %.2f \%%' % \
                     (Ec.mean()*100,(Ec.max()-Ec.min())*100)))
    else:
        print 'Pre-gamma  E (delta, mean):  %.2f  %.2f' % (100*(E_fit.max() - E_fit.min()), 100*E_fit.mean())
        print 'Post-gamma E (delta, mean):  %.2f  %.2f' % (100*(Ec.max() - Ec.min()), 100*Ec.mean())
    print 'Fit array:\n', d.fit_E_res*100
    return E_fit

def donly_fraction(d, E_th, fit=False, num_ph=False, gamma=1.):
    """Print donor-only fraction using a threshold or E fit data.
    
    Arguments:
        d (Data): the Data() object to analyze
        E_th (float): threshold on E to separate D-only and FRET bursts
        fit (bool): whether to use the fit data
        num_ph (bool): use number of photons (weighted burst size) to count
            D-only and FRET populations.
        gamma (float): gamma-factor to use to compute burst size
    
    Returns: 
        Per-ch donor-only fraction using threshold and fitting.
    """
    do_frac_fit = []
    if fit:
        if d.fit_E_res.shape[1] == 6:
            a, b  = d.fit_E_res[:, 2], d.fit_E_res[:, 5]
            do_frac_fit = a/(a + b)
        elif d.fit_E_res.shape[1] == 5:
            do_frac_fit = d.fit_E_res[:, -1]
        else:
            do_frac_fit = np.ones(d.nch)
            #raise ValueError('fit=True can be used only for 2-gaussian fits.')
        
    d_do = Sel(d, select_bursts.E, E2=E_th)
    d_fret = Sel(d, select_bursts.E, E1=E_th)
    if not num_ph:
        # D-only fraction as fraction of bursts
        do_frac_th = np.r_[[1.*do/(do+fret) for do, fret in zip(d_do.num_bu(), 
                                                          d_fret.num_bu())]]
    else:
        # D-only fraction as fraction of "corrected" photons
        do_frac_th = np.zeros(d.nch)
        for ich in xrange(d.nch):
            n_ph_do = select_bursts.get_burst_size(d_do, ich, gamma).sum()
            n_ph_fret = select_bursts.get_burst_size(d_fret, ich, gamma).sum()
            do_frac_th[ich] = n_ph_do/(n_ph_do + n_ph_fret)
    
    print '\n [ FRACTION of D-ONLY ]'
    print '\n     '+'  '.join(['CH%d']*8) % tuple(range(1,9)) + '  Mean Std.Dev'
    print 'Th:  '+' '.join(['%.2f']*10) % \
            (tuple(do_frac_th) + (do_frac_th.mean(), do_frac_th.std()*100))
    if fit:
        print 'Fit: '+' '.join(['%.2f']*10) % \
            (tuple(do_frac_fit) + (do_frac_fit.mean(), do_frac_fit.std()*100))
    return do_frac_th, do_frac_fit
```

Fitting here is a bit tricky due to the absence of two clearly resolved peaks (see FRET histograms below).

Here the fit paramenter $\alpha$ ('`a`', coefficienet of d-only component) is less constrained (0.08 .. 0.25) resulting
in a slightly lower fit because most CH collapse to `a=0.08` (low donor-only population):

```python
fit_type = '2gauss'
hist_weights = 'size'

gauss1_bounds = [(-0.01, 0.01), (0.01, 0.05), (0.08, 0.25)]
gauss2_bounds = [(0.05,0.3), (0.01,0.12), (0.25, 0.92)]

mu10, sig10, mu20, sig20, a10 = [0, 0.03, 0.6, 0.08, 0.2]
a20 = (1 - a10)

fret_fit_kwargs22 = dict(bins = fret_fit_bins,
                       #method = 'SLSQP',
                       method = 'L-BFGS-B',
                       verbose=False)

if fit_type == '2gauss':
    fret_fit_kwargs22['p0'] = [mu10, sig10, mu20, sig20, a10]
    fret_fit_kwargs22['bounds'] = gauss1_bounds[:2] + gauss2_bounds[:2] + gauss1_bounds[2:3]
elif fit_type == '2gauss_fun':
    fret_fit_kwargs22['p0'] = [mu10, sig10, a10, mu20, sig20, a20]
    fret_fit_kwargs22['bounds'] = gauss1_bounds + gauss2_bounds

dfs22 = Sel(d22, select_bursts.size, th1=30, gamma1=gamma1_sel)
dfs22t = Sel(d22, select_bursts.topN_nda, N=300, gamma1=gamma1_sel)
dfs22.fit_E_generic(fit_fun=fret_fit_func[fit_type], 
                    weights=hist_weights, gamma=gamma_fit, **fret_fit_kwargs22)

E_pr_fret.loc[data_id] = print_fit(dfs22, gamma_post)
do_frac_th22, _ = donly_fraction(dfs22, E_th=0.04, fit=True)
```

```python
fit_type = '2gauss_fun'
hist_weights = 'size'

gauss1_bounds = [(-0.01, 0.01), (0.01, 0.05), (0.15, 0.17)]
gauss2_bounds = [(0.05,0.3), (0.01,0.12), (0.83, 0.85)]

mu10, sig10, mu20, sig20, a10 = [0, 0.03, 0.6, 0.08, 0.2]
a20 = (1 - a10)

fret_fit_kwargs22 = dict(bins = fret_fit_bins,
                       #method = 'SLSQP',
                       method = 'L-BFGS-B',
                       verbose=False)

if fit_type == '2gauss':
    fret_fit_kwargs22['p0'] = [mu10, sig10, mu20, sig20, a10]
    fret_fit_kwargs22['bounds'] = gauss1_bounds[:2] + gauss2_bounds[:2] + gauss1_bounds[2:3]
elif fit_type == '2gauss_fun':
    fret_fit_kwargs22['p0'] = [mu10, sig10, a10, mu20, sig20, a20]
    fret_fit_kwargs22['bounds'] = gauss1_bounds + gauss2_bounds

dfs22 = Sel(d22, select_bursts.size, th1=30, gamma1=gamma1_sel)
dfs22t = Sel(d22, select_bursts.topN_nda, N=300, gamma1=gamma1_sel)
dfs22.fit_E_generic(fit_fun=fret_fit_func[fit_type],
                    weights=hist_weights, gamma=gamma_fit, **fret_fit_kwargs22)

E_fit22 = print_fit(dfs22, gamma_post)
do_frac_th22, _ = donly_fraction(dfs22, E_th=0.04, fit=True)
```

```python
def hist_fret_nb(i, b, d, bins=None, binw=0.02,
        no_text=False, normed=False, weights=None, gamma=1., ax=None, 
        noylabel=False, **kwargs):
    """Plot the FRET histogram and optionally the fitted model
    """
    if bins is None: bins = r_[-0.2:1.2:binw]
    if ax is None: ax = plt.subplot(111)
    plot_style = dict(histtype='step', lw=1.2,
                      #edgecolor='white'
                      )
    plot_style.update(**kwargs)  # kwargs overwrite plot_style
    if weights is not None:
        w = fret_fit.get_weights(d.nd[i], d.na[i], weights, gamma=gamma)
        w *= w.size/w.sum()
        plot_style.update(weights=w)
    H = ax.hist(1.*d.E[i], bins=bins, normed=normed, **plot_style)
    plt.xlabel('FRET Efficiency')
    if not noylabel:
        plt.ylabel('# Bursts')
        if normed: plt.ylabel('PDF')

def mhist_fret(d, **kwargs):
    for ich in xrange(d.nch):
        hist_fret_nb(ich, d.mburst[ich], d, **kwargs)
            
def get_empty_ax(fig, bgcolor='none'):
    ax = fig.add_subplot(111)
    ax.set_zorder(-1)
    ax.set_axis_bgcolor(bgcolor)
    ax.grid(False)
    #ax.set_frame_on(False)
    ax.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
    return ax
```
