# Stoichiometry fit

```python
weights_name = 'size'
weights = bl.fret_fit.get_weights(ds_fret.nd[0], ds_fret.na[0], weights=weights_name, gamma=1)
S = ds_fret.S[0]
```

As a comparison, let's fit of the histogram (curve-fit):

```python
S_fit = bl.gaussian_fit_hist(S, mu0=0.5, sigma0=0.1, bins=np.r_[-0.1:1.1:0.005], weights=weights)
S_fit
```

The Maximum likelihood fit for a Gaussian population is the mean:

```python
S_fit = (S.mean(), S.std())
S_fit
```

Computing the weighted mean and weighted standard deviation we get:

```python
S_mean = np.dot(weights, S)/weights.sum()
S_std_dev = np.sqrt(
        np.dot(weights, (S - S_mean)**2)/weights.sum())
S_fit = [S_mean, S_std_dev]
S_fit
```

```python
dplot(ds_fret, hist_S, normed=True, weights=weights_name)
x = np.arange(-0.1, 1.1, 0.005)
plot(x, bpl.normpdf(x, *S_fit), 'k')
```
